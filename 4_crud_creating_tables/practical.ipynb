# CRUD Creating Tables

## Teenage Movies

### 1. Use a SELECT statement to create a table called teenage_movies
### 2. Take the following columns from the film table: title, description, release_year and rating
### 3. Only SELECT movies which don't have the ratings, R or NC-17

```python
CREATE TABLE teenage_movies AS (
    SELECT title,
            description,
            release_year,
            rating
    FROM   film
    WHERE  NOT rating = 'R' OR
            NOT rating = 'NC-17'
);
```

## Years Payments

### 1. Use a SELECT statement to create a table called payments_this_year
### 2. SELECT all columns from the payment table
### 3. Only take the payments from the year 2007 

```python
CREATE TABLE payments_this_year AS (
    SELECT *
    FROM payment
    WHERE payment_date > '2007-01-01'::DATE
);
```

```python
execute_query('SELECT * FROM payments_this_year;')
```
