```python
df.dropna(inplace=True)
df.isna().sum()
```

# 5. For the numerical values, obtain the mean average value

```python
# Filter the DataFrame to select only numerical columns
numerical_df = df.select_dtypes(include=['number'])

# Calculate the mean for the numerical columns
numerical_df.mean()
```

# 6. Observe the Job column. How many categories can you find? What is the most frequent?

```python
# Count the number of unique categories in the "Job" column
# nunique counts number of unique categories in specified column
unique_categories = df['Job'].nunique()

# Find the most frequent category
# mode(): used to find the most frequent category.
# returns a `pandas series` with one or more values and we access the first value using [0] to get the most frequent category
most_frequent_job = df['Job'].mode()[0]

# Print the results
print(f"Number of unique categories in 'Job' column: {unique_categories}")
print(f"Most frequent job category: {most_frequent_job}")
```

# 7. Create a new column with the duration of each call in seconds

```python
call_start_times = pd.to_timedelta(df['CallStart'])
call_end_times = pd.to_timedelta(df['CallEnd'])

# Calculate the duration in seconds
df['CallDurationInSec'] = (call_end_times - call_start_times).dt.total_seconds()

# Print the DataFrame with the new 'CallDurationInSec' column
df
```

# 8. What is the average duration of each call?

```python
df['CallDurationInSec'].mean()
```
