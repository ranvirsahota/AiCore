## 1. Return the ids, title and release year of all films which have the category 'Animation'

```python
db.execute_query('''
                    SELECT 
                            film_id,
                            title,
                            release_year
                    FROM film
                    WHERE film_id IN (
                        SELECT film_id
                        FROM film_category
                        WHERE category_id = (
                            SELECT category_id
                            FROM category
                            WHERE name = 'Animation'
                        )
                    );
''')
```

## 2. Return the first name, last name, and email of all customers in Canada

```python
db.execute_query('''
                    SELECT first_name,
                        last_name,
                        email
                    FROM customer
                    WHERE address_id IN (
                        SELECT address_id
                        FROM address
                        WHERE city_id IN (
                            SELECT city_id
                            FROM city
                            WHERE country_id = (
                                SELECT country_id
                                FROM country
                                WHERE country = 'Canada'
                            )
                        )
                    );
''')
```

## 3. Return the titles of films with movies starting with A or I and are not in the Italian, French or Spanish Language (all films are in English, but write the query as the practical requests)

```python
db.execute_query('''
                    SELECT title
                    FROM film
                    WHERE title LIKE 'A%%' OR title LIKE 'I%%'
                        AND language_id NOT IN (
                            SELECT language_id
                            FROM language
                            WHERE name IN ('Italian', 'French', 'Spanish')
                        );

''')
# DOUBLE '%' ONLY FOR SQLALCHEMY TO PARSE REGEX
```

## 4. Return the average film length per rating

```python
db.execute_query('''
                    SELECT *
                    FROM (
                        SELECT rating,
                        ROUND(AVG(length), 2) AS avg_length
                        FROM film
                        GROUP BY rating
                    ) AS avg_length_per_rating;
''')
```

## 5. Find the average number of sales per day for each staff

```python
db.execute_query('''
                    SELECT *
                    FROM (
                        SELECT 
                                staff_id,
                                ROUND(COUNT(amount) / EXTRACT('day' FROM (MAX(payment_date) - MIN(payment_date)))) AS staff_sales_per_day
                        FROM payment
                        GROUP BY staff_id
                    ) AS avg_sales_per_day_for_each_staff
''')
```
