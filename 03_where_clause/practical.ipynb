## 1. Return 10 films who's lengths are under 120 mins

```python
execute_query('''
                SELECT film_id, title, length
                FROM film
                WHERE length < 120
                LIMIT 10;
            ''').set_index('film_id')
```

## 2. Return the 10 longest films who's rating are G (Hint: to search by string we need to wrap our query in SINGLE quotes (e.g. 'G'))

```python
execute_query('''
                SELECT film_id, title, rating
                FROM film
                WHERE rating = 'G'
                LIMIT 10;
            ''')
```

## 3. Return all transactions where payment has been above $10

```python
execute_query('''
                SELECT payment_id, rental_id, amount
                FROM payment
                WHERE amount > 10;
            ''')
```

## 4. Return the (replacement) cost per minute of every movie


```python
print(execute_query('''
                SELECT film_id, 
                       title,
                       rental_rate,
                       length,
                       replacement_cost,
                       (replacement_cost/length) AS replacement_cost_per_minute
                FROM FILM;
            ''').set_index('film_id'))

execute_query('SELECT SUM(replacement_cost/length) total_replacement_cost_per_minute FROM FILM')
```

## 5. Return the top 10 most expensive films to rent, based on the rental rate per hour of the movie

```python
execute_query('''
                SELECT film_id,
                       title,
                       rental_rate,
                       length
                FROM film
                ORDER BY rental_rate ASC;
              ''')
```
