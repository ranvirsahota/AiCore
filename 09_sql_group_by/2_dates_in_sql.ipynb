# SQL GROUP BY - USING Aggregated Functions - PRACTICALS

## 1. List the last names of actors, as well as how many actors have that last name.

```python
db.execute_query('''
                    SELECT
                            last_name,
                            COUNT(last_name) AS occurrence
                    FROM actor
                    GROUP BY last_name;
                ''')
```

## 2. How many customers are there per store? Return the number of customers and the store id.

```python
db.execute_query('''
					SELECT
							store_id,
							COUNT(*) AS total_customers
					FROM customer
					GROUP BY store_id;
				''')
```

## 3. When was the earliest order placed per customer? Return the customer ids, the rental rates, and the names

```python
db.execute_query('''
					SELECT 
							DISTINCT ON (customer.customer_id)
							customer.customer_id,
							customer.first_name,
							customer.last_name,
							film.rental_rate,
							MIN(rental.rental_date) AS earliest_order_date
					FROM customer
					INNER JOIN rental ON rental.customer_id = customer.customer_id
					INNER JOIN inventory ON inventory.inventory_id = rental.inventory_id
					INNER JOIN film ON film.film_id = inventory.film_id
					GROUP BY
						customer.customer_id,
						customer.first_name,
						customer.last_name,
						film.rental_rate
					ORDER BY
						customer.customer_id ASC, earliest_order_date ASC;
				''')
```

## 4. What was the largest order placed per customer? Return the customer ids and the amounts

```python
db.execute_query('''
					SELECT 
							customer_id,
							MAX(amount) AS largest_amount_per_customer
					FROM payment
					GROUP BY customer_id
					ORDER BY largest_amount_per_customer DESC;
				''')
```
