## 1. Find the unique special features of films

## 2. Which 3 days of the week are most profitable for the business?

```python
db.execute_query('''
					SELECT
							EXTRACT('dow' from payment_date) AS day_of_the_week,
							SUM(amount) AS total_amount
					FROM payment
					GROUP BY day_of_the_week
					ORDER BY total_amount DESC
					LIMIT 3;
				''')
```

## 3. Return the total sales per day, along with the number of movies rented for that day

```python
db.execute_query('''
					SELECT
							EXTRACT('day' FROM payment_date) AS day_of_the_week,
							COUNT(payment.payment_date) AS sale_per_day,
							COUNT(rental.rental_date) AS total_movies_rented
							
							
					FROM payment
					INNER JOIN rental ON rental.rental_id = payment.rental_id
					GROUP BY day_of_the_week
					ORDER BY day_of_the_week ASC;
				''')
```

## 4. Find the id's of all customers who have spent over $100 over the course of their membership

```python
db.execute_query('''
					SELECT 
						customer_id,
						SUM(amount) AS total_spent
					FROM payment
					GROUP BY customer_id
					HAVING SUM(amount) > 100
					ORDER BY total_spent DESC;
				''')
```
